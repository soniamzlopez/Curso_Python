```python
#Inserta aquí tu código

# 9,99€ PRECIO TOTAL
# 21% IVA
# 121% -> 9,99€
# 100% base imponible -> 8,25€
# 21% IVA -> 1,73€

def apply_iva():
  base_imponible = 15
  base_imponible * 121 / 100

# Calcula la función en función de la base imponible
def apply_iva(base_imponible):
  return base_imponible * 121 / 100

# El IVA no siempre es 21%
def apply_iva(base_imponible, porcentaje_iva):
  iva = 100 + porcentaje_iva
  return base_imponible * iva / 100
```

### Ejercicio 6
** Escribe una función que aplica la función cuadrado() a todos los elementos de una lista.**

    Parámetros:
        funcion: Recibe la función a aplicar.
        lista: Es una lista con valores que se pasarán como argumentos a funcion.
    Devuelve:
        Una lista con el resultado de aplicar la función a los valores de la lista.

```python
def cuadrado(n):
    return n ** 2
```

```python
#Inserta aquí tu código
lista = [1,2,3,4,5]


resultado = []
for numero in lista:
  resultado.append(numero ** 2)
print(resultado)

# Meter en una función que va a recibir una lista como parámetro de entrada
# y la función que queramos aplicarle (cuadrado, logaritmo...)
def ejercicio6(lista, funcion):
  resultado = []
  for numero in lista:
    resultado.append(funcion(numero))
  return resultado
```

```python
# Esto sería lo mismo que list(map(funcion, lista)) / lista(map(cuadrado), [1,2,3,4,6])
# Con lambda también se puede: list(map(lambda numero : numero ** 2, [1,2,3,4,6]))
ejercicio6([1,2,3,4,6], cuadrado)
```

```python
# Otra manera de hacerlo aunque tarda más tiempo en correr la función, porque hay pocos datos
rdd_lista = sc.parallelize([1,2,3,4,6])
rdd_lista.map(lambda numero : numero ** 2).collect()
```
